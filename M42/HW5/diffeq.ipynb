```python
t = linspace(-1, 1, 200)
a = (math.pi * np.exp(1*t)) / ((np.exp(1*t))-1 + math.pi)
b = (math.pi * np.exp(2*t)) / ((np.exp(2*t))-1 + math.pi)
c = (math.pi * np.exp(3*t)) / ((np.exp(3*t))-1 + math.pi)
d = (math.pi * np.exp(4*t)) / ((np.exp(4*t))-1 + math.pi)
e = (math.pi * np.exp(5*t)) / ((np.exp(5*t))-1 + math.pi)

plt.plot(t, a, 'red') # r is 1
plt.plot(t, b, 'blue') # r is 2
plt.plot(t, c, 'green') # r is 3
plt.plot(t, d, 'purple') # r is 4
plt.plot(t, e, 'yellow') # r is 5
plt.show()
```
